# Data Preprocessing

```python
dropout_df.replace('NR',0,inplace=True)
```

```python
dropout_df.replace('Uppe_r_Primary',0,inplace=True)
```

```python
# Converting datatype of column (Primary Boys and Girls) from object to float.

dropout_df['Primary_Boys'] = dropout_df['Primary_Boys'].apply(lambda x: float(x))
dropout_df['Primary_Girls'] = dropout_df['Primary_Girls'].apply(lambda x: float(x))
```

```python
# Converting datatype of column (Upper Primary Boys and Girls) from object to float.

dropout_df['Upper Primary_Boys'] = dropout_df['Upper Primary_Boys'].apply(lambda x: float(x))
dropout_df['Upper Primary_Girls'] = dropout_df['Upper Primary_Girls'].apply(lambda x: float(x))
```

```python
# Converting datatype of column (Secondary Boys and Girls) from object to float.

dropout_df['Secondary _Boys'] = dropout_df['Secondary _Boys'].apply(lambda x: float(x))
dropout_df['Secondary _Girls'] = dropout_df['Secondary _Girls'].apply(lambda x: float(x))
```

```python
# Converting datatype of column (HrSecondary Boys and Girls) from object to float.

dropout_df['HrSecondary_Boys'] = dropout_df['HrSecondary_Boys'].apply(lambda x: float(x))
dropout_df['HrSecondary_Girls'] = dropout_df['HrSecondary_Girls'].apply(lambda x: float(x))
```

```python
# Converting datatype of column ( Total of Primary, Upper Primary, Secondary and HrSecondary Boys 
# and Girls) from object to float.

# Primary Total
dropout_df['Primary_Total'] = dropout_df['Primary_Total'].apply(lambda x: float(x))

# Upper Primary Total
dropout_df['Upper Primary_Total'] = dropout_df['Upper Primary_Total'].apply(lambda x: float(x))

# Secondary Total
dropout_df['Secondary _Total'] = dropout_df['Secondary _Total'].apply(lambda x: float(x))

#HrSecondary Total
dropout_df['HrSecondary_Total'] = dropout_df['HrSecondary_Total'].apply(lambda x: float(x))
```

```python
#Check info() of dataframe
dropout_df.info()
```

```python
dropout_df.replace('Arunachal  Pradesh','Arunachal Pradesh',inplace=True)

dropout_df.replace('Madhya  Pradesh','Madhya Pradesh',inplace=True)

dropout_df.replace('Tamil  Nadu','Tamil Nadu',inplace=True)
```

# Outliers Removal

```python
dropout_df['Primary_Boys']=dropout_df['Primary_Boys'].clip(lower=dropout_df['Primary_Boys'].quantile(0.05), upper=dropout_df['Primary_Boys'].quantile(0.95))
dropout_df['Primary_Girls']=dropout_df['HrSecondary_Boys'].clip(lower=dropout_df['HrSecondary_Boys'].quantile(0.05), upper=dropout_df['HrSecondary_Boys'].quantile(0.95))

dropout_df['Upper Primary_Boys']=dropout_df['Upper Primary_Boys'].clip(lower=dropout_df['Upper Primary_Boys'].quantile(0.05), upper=dropout_df['Upper Primary_Boys'].quantile(0.95))
dropout_df['Upper Primary_Girls']=dropout_df['Upper Primary_Girls'].clip(lower=dropout_df['Upper Primary_Girls'].quantile(0.05), upper=dropout_df['Upper Primary_Girls'].quantile(0.95))

dropout_df['Secondary _Boys']=dropout_df['Secondary _Boys'].clip(lower=dropout_df['Secondary _Boys'].quantile(0.05), upper=dropout_df['Secondary _Boys'].quantile(0.95))
dropout_df['Secondary _Girls']=dropout_df['Secondary _Girls'].clip(lower=dropout_df['Secondary _Girls'].quantile(0.05), upper=dropout_df['Secondary _Girls'].quantile(0.95))

dropout_df['HrSecondary_Boys']=dropout_df['HrSecondary_Boys'].clip(lower=dropout_df['HrSecondary_Boys'].quantile(0.05), upper=dropout_df['HrSecondary_Boys'].quantile(0.95))
dropout_df['HrSecondary_Girls']=dropout_df['HrSecondary_Girls'].clip(lower=dropout_df['HrSecondary_Girls'].quantile(0.05), upper=dropout_df['HrSecondary_Girls'].quantile(0.95))
```

# Conclusion

1. According to above Analysis it came to know that most of the students take dropout in year 2013-2014 (36%) duration and after that in year 2014-2015 (34%) duration. 

2. One more information is came to know from above Analysis is that, in the education stage (Primary and Upper Primary), 'Girls' or 'Female' students are take the most dropout. And in the education stage (Secondary and Hr. Secondary), 'Boys' or 'Male' students are take the most dropout.
