# Data Preprocessing

```python
# merge both the dataframe and make a single one named 'df'
df = pd.merge(boys_df,girls_df,how='outer')
```

```python
df.head()
```

```python
df.describe()
```

```python
df.info()
```

```python
df.columns
```

```python
df.isna().sum()
```

```python
df['State_UT'].nunique()
```

```python
df['State_UT'].unique()
```

```python
# Make a single column for Primary Class
df['Primary'] = df[['Primary_Only','Primary_with_U_Primary']].mean(axis=1)

# Make a single column for Upper Primary Class
df['Upper_Primary'] = df[['U_Primary_Only','U_Primary_With_Sec','Primary_with_U_Primary_Sec']].mean(axis=1)

# Make a single column for Secondary Class
df['Secondary'] = df[['Sec_Only','U_Primary_With_Sec_HrSec']].mean(axis=1)

# Make a single column for Hr Secondary Class
df['Hr_Secondary'] = df[['HrSec_Only','Sec_with_HrSec.']].mean(axis=1)

# Make a single column for All school class
df['All_Schools'] = df[['All Schools','Primary_with_U_Primary_Sec_HrSec']].mean(axis=1)
```

```python
df.drop(['Primary_Only', 'Primary_with_U_Primary','Primary_with_U_Primary_Sec_HrSec', 'U_Primary_Only',
       'U_Primary_With_Sec_HrSec', 'Primary_with_U_Primary_Sec','U_Primary_With_Sec', 'Sec_Only', 
         'Sec_with_HrSec.', 'HrSec_Only','All Schools'],axis=1,inplace=True)
```

```python
df.describe()
```

# Exploratory Data Analysis

```python
plt.figure(figsize=(20,20))
sns.pairplot(data=df)
```

```python
sns.countplot(data=df,x='year')
```

```python
# Distribution of Years in Percentage

year_dr1 = df[['Primary','Upper_Primary','Secondary','Hr_Secondary','All_Schools']][df['year']=='2013-14'].sum()
year_dr2 = df[['Primary','Upper_Primary','Secondary','Hr_Secondary','All_Schools']][df['year']=='2014-15'].sum()
year_dr3 = df[['Primary','Upper_Primary','Secondary','Hr_Secondary','All_Schools']][df['year']=='2015-16'].sum()
```

```python
years_duration_data = [year_dr1.sum(), year_dr2.sum(), year_dr3.sum()]
years_duration = ['2013-2014', '2014-2015', '2015-2016']

plt.pie(years_duration_data, labels=years_duration, autopct='%.0f%%',startangle=90)
```

```python
# Distribution of Classes in Percentage

pri_class = df['Primary'].sum()
upp_pri_class = df['Upper_Primary'].sum()
secondary_class = df['Secondary'].sum()
hr_sec_class = df['Hr_Secondary'].sum()
```

```python
class_data = [pri_class, upp_pri_class, secondary_class, hr_sec_class]
class_label = ['Primary', 'Upper Primary', 'Secondary', 'Hr. Secondary']

plt.pie(class_data, labels=class_label, autopct='%.0f%%',startangle=90)
```

Now, we have one single dataset which is easy to process. Our dataset is much good because they not contain any null values, missing values and etc. But, thier is one problem that the number of columns is more and they labeled in some confusing manner. So, we need to make a single column for a particular class like:- Primary, Upper Primary, Secondary, Hr. Secondary and All Schools (those who not categories in any class). After all Data Preprocessing, we need to do some exploratory data analysis (EDA) to make some usefull insights from our data. We done some Univariate Data Analysis and Bivariate Data Analysis for better understanding of data.
