```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
import warnings
warnings.filterwarnings('ignore')
```

```python
from sklearn.datasets import make_blobs
```

```python
data = make_blobs(n_samples=200,n_features=2,centers=4,cluster_std=1.8,random_state=101)
```

```python
plt.figure(figsize=(12,6))
plt.scatter(data[0][:,0],data[0][:,1],c=data[1],cmap='rainbow')
```

```python
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(n_clusters=3)
```

```python
kmeans.fit(data[0])
```

```python
kmeans.cluster_centers_
```

```python
kmeans.labels_
```

```python
fig , (axs1,axs2) = plt.subplots(2,1, sharey=True,figsize = (12,10))

axs1.set_title('K Means')
axs1.scatter(data[0][:,0],data[0][:,1],c=kmeans.labels_,cmap='rainbow')

axs2.set_title('Original')
axs2.scatter(data[0][:,0],data[0][:,1],c=data[1],cmap='rainbow')
```
